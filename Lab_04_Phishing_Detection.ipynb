```python
import re
from urllib.parse import urlparse

# Function to analyze the URL
def check_phishing_url(url):

    score = 0
    indicators = []

    # Add http:// if no scheme is provided
    if not url.startswith(("http://", "https://")):
        url = "http://" + url

    parsed_url = urlparse(url)
    domain = parsed_url.netloc.lower()

    # 1. Check for IP address instead of domain name
    ip_pattern = r"^\d{1,3}(\.\d{1,3}){3}$"

    if re.match(ip_pattern, domain):
        score += 2
        indicators.append("Uses an IP address instead of a domain name")

    # 2. Check for '@' symbol
    if "@" in url:
        score += 2
        indicators.append("Contains '@' symbol")

    # 3. Check for excessive hyphens
    if domain.count("-") >= 2:
        score += 1
        indicators.append("Contains multiple hyphens in the domain")

    # 4. Check for very long URL
    if len(url) > 75:
        score += 1
        indicators.append("URL is unusually long")

    # 5. Check for many subdomains
    if domain.count(".") >= 3:
        score += 1
        indicators.append("Contains many subdomains")

    # 6. Check for suspicious keywords
    suspicious_keywords = [
        "login",
        "verify",
        "account",
        "secure",
        "update",
        "banking",
        "password",
        "confirm"
    ]

    for keyword in suspicious_keywords:
        if keyword in url.lower():
            score += 1
            indicators.append(
                f"Contains suspicious keyword: {keyword}"
            )

    # 7. Check whether HTTPS is used
    if not url.startswith("https://"):
        score += 1
        indicators.append("Does not use HTTPS")

    # Display results
    print("\n--- URL ANALYSIS RESULT ---")
    print("URL:", url)
    print("Domain:", domain)

    if indicators:
        print("\nSuspicious Indicators Found:")

        for indicator in indicators:
            print("-", indicator)
    else:
        print("\nNo common phishing indicators found.")

    print("\nRisk Score:", score)

    # Classification
    if score >= 4:
        print("Classification: LIKELY PHISHING URL")

    elif score >= 2:
        print("Classification: SUSPICIOUS URL")

    else:
        print("Classification: LIKELY SAFE URL")


# Get URL from user
url = input("Enter the URL to check: ")

# Analyze the URL
check_phishing_url(url)
```
