```python
import hashlib

# Function to calculate SHA-256 hash of a file
def calculate_hash(filename):
    sha256 = hashlib.sha256()

    with open(filename, "rb") as file:
        while True:
            data = file.read(4096)

            if not data:
                break

            sha256.update(data)

    return sha256.hexdigest()


# Create an original file
filename = "sample.txt"

with open(filename, "w") as file:
    file.write("This is the original secure data.")

print("Original file created successfully.")


# Calculate original hash
original_hash = calculate_hash(filename)

print("\nOriginal File Hash:")
print(original_hash)


# Simulate file tampering
with open(filename, "a") as file:
    file.write("\nUnauthorized modification by attacker.")

print("\nFile has been modified.")


# Calculate hash after modification
current_hash = calculate_hash(filename)

print("\nCurrent File Hash:")
print(current_hash)


# Compare the hashes
if original_hash == current_hash:
    print("\nFile is safe. No tampering detected.")
else:
    print("\nWARNING: File tampering detected!")
```
