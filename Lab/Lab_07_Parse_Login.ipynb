```python
# -----------------------------------------
# Login Log Parser
# -----------------------------------------

# Sample login log
log_data = """2026-07-16 10:00:10 192.168.1.10 SUCCESS
2026-07-16 10:01:25 192.168.1.15 FAILED
2026-07-16 10:02:40 192.168.1.20 FAILED
2026-07-16 10:03:18 192.168.1.15 FAILED
2026-07-16 10:04:22 192.168.1.25 SUCCESS
2026-07-16 10:05:30 192.168.1.20 FAILED
2026-07-16 10:06:45 192.168.1.15 SUCCESS
2026-07-16 10:07:50 192.168.1.30 FAILED
2026-07-16 10:08:15 192.168.1.20 FAILED
"""

# Create the log file
with open("login_log.txt", "w") as file:
    file.write(log_data)

print("Login log file created successfully.\n")

failed_attempts = {}

# Read the log file
with open("login_log.txt", "r") as file:

    for line in file:

        parts = line.split()

        ip = parts[2]
        status = parts[3]

        if status == "FAILED":

            failed_attempts[ip] = failed_attempts.get(ip, 0) + 1

print("=" * 45)
print("FAILED LOGIN ATTEMPTS PER IP")
print("=" * 45)

for ip, count in failed_attempts.items():
    print(f"{ip:<15} : {count}")

print("=" * 45)
```
