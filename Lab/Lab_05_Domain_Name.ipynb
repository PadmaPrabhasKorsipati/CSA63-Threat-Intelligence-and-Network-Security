```python
import socket

# Get number of domains from the user
n = int(input("Enter the number of domains: "))

domains = []

# Get domain names from the user
print("\nEnter the domain names:")

for i in range(n):
    domain = input(f"Domain {i + 1}: ")
    domains.append(domain)


# Display heading
print("\n--- DOMAIN NAME RESOLUTION RESULTS ---\n")


# Resolve each domain name
for domain in domains:

    try:
        # Resolve domain name to IP address
        ip_address = socket.gethostbyname(domain)

        print(f"Domain Name : {domain}")
        print(f"IP Address  : {ip_address}")
        print("-" * 40)

    except socket.gaierror:

        print(f"Domain Name : {domain}")
        print("IP Address  : Unable to resolve domain")
        print("-" * 40)


print("\nDomain resolution completed.")
```
