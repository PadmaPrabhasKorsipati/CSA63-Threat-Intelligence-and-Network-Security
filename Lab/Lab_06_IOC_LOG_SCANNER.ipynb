```python
# ---------------------------------------------
# Threat Intelligence - IoC Matching
# ---------------------------------------------

# List of known malicious IP addresses (IoCs)
malicious_ips = {
    "192.168.1.10",
    "45.33.32.156",
    "103.21.244.15",
    "8.8.8.8"
}

# Sample network traffic log
traffic_log = [
    {"Time": "10:00", "Source_IP": "192.168.1.5", "Destination_IP": "172.16.0.10"},
    {"Time": "10:05", "Source_IP": "45.33.32.156", "Destination_IP": "192.168.1.20"},
    {"Time": "10:10", "Source_IP": "10.0.0.25", "Destination_IP": "8.8.8.8"},
    {"Time": "10:15", "Source_IP": "192.168.1.10", "Destination_IP": "172.16.0.15"},
    {"Time": "10:20", "Source_IP": "192.168.1.30", "Destination_IP": "192.168.1.40"}
]

print("=" * 60)
print("THREAT INTELLIGENCE - IOC MATCHING")
print("=" * 60)

matches = 0

for record in traffic_log:

    src = record["Source_IP"]
    dst = record["Destination_IP"]

    if src in malicious_ips or dst in malicious_ips:

        matches += 1

        print("\n⚠ ALERT : Malicious IP Detected")
        print("-------------------------------------")
        print("Time            :", record["Time"])
        print("Source IP       :", src)
        print("Destination IP  :", dst)

print("\n" + "=" * 60)

if matches == 0:
    print("No malicious IP addresses detected.")
else:
    print(f"Total malicious matches found : {matches}")

print("=" * 60)
```
