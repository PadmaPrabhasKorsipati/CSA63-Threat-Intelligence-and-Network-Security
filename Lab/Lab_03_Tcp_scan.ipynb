```python
import socket
import threading
import time

# -------------------------------------------------
# STEP 1: Start a temporary TCP server on port 8000
# -------------------------------------------------

HOST = "127.0.0.1"
TEST_PORT = 8000

def start_test_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    # Allow the port to be reused
    server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

    # Bind the server to localhost and port 8000
    server_socket.bind((HOST, TEST_PORT))

    # Start listening for connections
    server_socket.listen(5)

    print(f"Test server started on {HOST}:{TEST_PORT}")

    # Keep the server running
    while True:
        try:
            client_socket, address = server_socket.accept()
            client_socket.close()
        except:
            break


# Start the test server in the background
server_thread = threading.Thread(
    target=start_test_server,
    daemon=True
)

server_thread.start()

# Wait for the server to start
time.sleep(1)


# -------------------------------------------------
# STEP 2: TCP PORT SCANNER
# -------------------------------------------------

target = "127.0.0.1"
start_port = 7995
end_port = 8005

print(f"\nScanning {target}")
print(f"Port range: {start_port} to {end_port}\n")

open_ports = []

# Scan each port
for port in range(start_port, end_port + 1):

    # Create a TCP socket
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    # Set connection timeout
    sock.settimeout(0.5)

    # Try to connect to the port
    result = sock.connect_ex((target, port))

    # If result is 0, the port is open
    if result == 0:
        print(f"Port {port} is OPEN")
        open_ports.append(port)
    else:
        print(f"Port {port} is CLOSED")

    # Close the socket
    sock.close()


# -------------------------------------------------
# STEP 3: DISPLAY FINAL RESULT
# -------------------------------------------------

print("\nScan completed.")

if open_ports:
    print("Open ports:", open_ports)
else:
    print("No open ports found.")
```
